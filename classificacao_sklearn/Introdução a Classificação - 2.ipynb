```python
import pandas as pd
url_tracking = "https://gist.githubusercontent.com/guilhermesilveira/b9dd8e4b62b9e22ebcb9c8e89c271de4/raw/c69ec4b708fba03c445397b6a361db4345c83d7a/tracking.csv"
dados = pd.read_csv(url_tracking)
```

```python
x = dados[["inicial",	"palestras", "contato", "patrocinio"]]
```

```python
y = dados[["comprou"]]
```

```python
from sklearn.model_selection import train_test_split

SEED = 67
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED)
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

print(f'Treinaremos com {len(treino_x)} elementos')
print(f'Testaremos com {len(teste_x)} elementos')
```

```python
previsoes = modelo.predict(teste_x)
```

```python
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(teste_y, previsoes) * 100
print(f'Acurácia: {acuracia:.2f} %')
```
